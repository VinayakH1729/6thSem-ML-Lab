```python
#Handling categorical variables
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
```

```python
#Gender and class are categorical as they are of object datatype
print(diabetes["Gender"].value_counts())
print(diabetes["CLASS"].value_counts())
```

```python
diabetes["Gender"] = diabetes["Gender"].replace("f", "F")
updated_file_path = "/content/Dataset of Diabetes .csv"  # Replace with the desired path
diabetes.to_csv(updated_file_path, index=False)
```

```python
diabetes["Gender"].value_counts()
```

```python
ordinal_encoder = OrdinalEncoder(categories=[["M", "F"]])
diabetes["Gender_Encoded"] = ordinal_encoder.fit_transform(diabetes[["Gender"]])
```

```python
diabetes["CLASS"] = diabetes["CLASS"].replace(["N ","Y "],["N","Y"])
updated_file_path = "/content/Dataset of Diabetes .csv"  # Replace with the desired path
diabetes.to_csv(updated_file_path, index=False)
```

```python
diabetes['CLASS'].value_counts()
```

```python
onehot_encoder = OneHotEncoder()
encoded_data = onehot_encoder.fit_transform(diabetes[["CLASS"]])
encoded_array = encoded_data.toarray()
encoded_df = pd.DataFrame(encoded_array, columns=onehot_encoder.get_feature_names_out(["CLASS"]))
df_encoded = pd.concat([diabetes, encoded_df], axis=1)

df_encoded.drop("Gender", axis=1, inplace=True)
df_encoded.drop("CLASS", axis=1, inplace=True)

print(df_encoded. head())
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from scipy import stats
```

```python
normalizer = MinMaxScaler()
df_encoded[['No_Pation']] = normalizer.fit_transform(df_encoded[['No_Pation']])
df_encoded.head()
```

```python
scaler = StandardScaler()
df_encoded[['AGE']] = scaler.fit_transform(df_encoded[['AGE']])
df_encoded.head()
```

```python
import numpy as np
```

```python
df_encoded_copy1=df_encoded
df_encoded_copy2=df_encoded
df_encoded_copy3=df_encoded

Q1 = df_encoded_copy1['No_Pation'].quantile(0.25)
Q3 = df_encoded_copy1['No_Pation'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_encoded_copy1['No_Pation'] = np.where(df_encoded_copy1['No_Pation'] > upper_bound, upper_bound,
                        np.where(df_encoded_copy1['No_Pation'] < lower_bound, lower_bound, df_encoded_copy1['No_Pation']))

print(df_encoded_copy1.head())
```

```python
df_encoded_copy2['No_Pation_zscore'] = stats.zscore(df_encoded_copy2['No_Pation'])
df_encoded_copy2['No_Pation'] = np.where(df_encoded_copy2['No_Pation_zscore'].abs() > 3, np.nan, df_encoded_copy2['No_Pation'])  # Replace outliers with NaN
print(df_encoded_copy2.head())
```

```python
df_encoded_copy3['No_Pation_zscore'] = stats.zscore(df_encoded_copy3['No_Pation'])
median_salary = df_encoded_copy3['No_Pation'].median()
df_encoded_copy3['No_Pation'] = np.where(df_encoded_copy3['No_Pation_zscore'].abs() > 3, median_salary, df_encoded_copy3['No_Pation'])
print(df_encoded_copy3.head())
```
